```python
import pandas as pd
```

```python
pparks = pd.read_csv("http://data-phl.opendata.arcgis.com/datasets/d52445160ab14380a673e5849203eb64_0.csv")
pparks
```

```python
meanps = pparks["ACREAGE"].mean()
meanps
```

```python
medianps = pparks["ACREAGE"].median()
medianps
```

```python
import pylab as plt
pparks['ACREAGE'].hist(bins=20, alpha=0.7)

# Overplot mean and median
plt.axvline(meanps, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(medianps, color='green', linestyle='dashed', linewidth=1, label='Median')

# Add labels and title
plt.xlabel('Park size in acres')
plt.ylabel('Number of parks')
plt.title('Distribution of Park Sizes')
plt.legend()

# Show the plot
plt.show()
```

```python
stdps = pparks["ACREAGE"].std()
stdps
```

```python
ax = pparks[pparks["ACREAGE"] < 10]
ax
```

```python
meanax = ax["ACREAGE"].mean()
meanax
```

```python
medianax = ax["ACREAGE"].median()
medianax
```

```python
ax['ACREAGE'].hist(bins=10, alpha=0.7)

# Overplot mean and median
plt.axvline(meanax, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(medianax, color='green', linestyle='dashed', linewidth=1, label='Median')

# Add labels and title
plt.xlabel('Park size in acres')
plt.ylabel('Number of parks')
plt.title('Distribution of Park Sizes')
plt.legend()

# Show the plot
plt.show()
```

```python
parent_parks = ax.groupby("PARENT_NAME")["ACREAGE"].sum()
parent_parks
```
